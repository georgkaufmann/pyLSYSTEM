# L-Systems: Branching examples
----
In this notebook, we draw branching examples of **L systems**.

```python
import libLSYSTEM
```

----
## Tree

```python
axiom = 'X'
rules = {'F': 'FF',
         'X': 'F-[[X]+X]+F[+FX]-X',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=6)
libLSYSTEM.lsystemPlot(word,length=0.2,angle=22.5,x=0,y=0,d=90)
```

----
## Leaf

```python
axiom = 'a'
rules = {'F': '>F<',
         'a': 'F[+x]Fb',
         'b': 'F[-y]Fa',
         'x': 'a',
         'y': 'b',
         'X': 'X',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+',
         '>': '>',
         '<': '<'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=20)
libLSYSTEM.lsystemPlot(word,length=0.02,angle=45,scale=1.36,x=-5,y=-5,d=90)
#libLSYSTEM.lsystemPlot(word,length=0.02,angle=22,scale=1.36,x=-5,y=-5,d=90)
```
