# L-Systems: Non-branching examples
----
In this notebook, we draw non-branching examples of **L systems**.

```python
import libLSYSTEM
```

----
## Koch island

```python
axiom = 'F-F-F-F'
rules = {'F': 'F-F+F+FF-F-F+F',
         'X': 'X',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=3)
libLSYSTEM.lsystemPlot(word,length=0.15,angle=90,x=-5,y=-5,d=0)
```

```python
axiom = '-F'
rules = {'F': 'F+F-F-F+F',
         'X': 'X',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=4)
libLSYSTEM.lsystemPlot(word,length=0.15,angle=90,x=-5,y=-5,d=90)
```

```python
axiom = 'F+F+F+F'
rules = {'F': 'F+f-FF+F+FF+Ff+FF-f+FF-F-FF-Ff-FFF',
         'f': 'ffffff',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=2)
libLSYSTEM.lsystemPlot(word,length=0.15,angle=90,x=-5,y=-5,d=90)
```

```python
axiom = 'F-F-F-F'
rules = {'F': 'FF-F-F-F-FF',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=4)
libLSYSTEM.lsystemPlot(word,length=0.15,angle=90,x=-5,y=-5,d=90)
```

----
## Koch snowflake

```python
axiom = 'F++F++F'
rules = {'F': 'F-F++F-F',
         'X': 'X',
         '[': '[',
         ']': ']',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=6)
libLSYSTEM.lsystemPlot(word,length=0.02,angle=60,x=-5,y=-5,d=0)
```

----
## Crystal

```python
axiom = 'F+F+F+F'
rules = {'F': 'FF+F++F+F',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=6)
libLSYSTEM.lsystemPlot(word,length=0.02,angle=90,scale=1.36,x=-5,y=-5,d=90)
```

----
## Sierpinski triangle

```python
axiom = 'F-G-G'
rules = {'F': 'F-G+F+G-F',
         'G': 'GG',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=6)
libLSYSTEM.lsystemPlot(word,length=0.02,angle=120,scale=1.36,x=-5,y=-5,d=180)
```

```python
axiom = 'F'
rules = {'F': 'G-F-G',
         'G': 'F+G+F',
         '-': '-',
         '+': '+'}

word  = libLSYSTEM.lsystemIterate(axiom,rules,iter=7)
libLSYSTEM.lsystemPlot(word,length=0.02,angle=60,scale=1.36,x=-5,y=-5,d=0)
```
