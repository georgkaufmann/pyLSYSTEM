# Sierpinski triangle
----

In this notebook, we create the **Sierpinski triangle**.

~~~
From Wikipedia, the free encyclopedia:

"The Sierpiński triangle (sometimes spelled Sierpinski), also called the Sierpiński gasket or Sierpiński sieve, is a fractal attractive fixed set with the overall shape of an equilateral triangle, subdivided recursively into smaller equilateral triangles. Originally constructed as a curve, this is one of the basic examples of self-similar sets—that is, it is a mathematically generated pattern that is reproducible at any magnification or reduction. It is named after the Polish mathematician Wacław Sierpiński, but appeared as a decorative pattern many centuries before the work of Sierpiński. "
~~~

```python
import numpy as np
import matplotlib.pyplot as plt
```

We will learn to call a function **iteratively**. We use the **removal** method:

The Sierpinski triangle may be constructed from an equilateral triangle by 
repeated removal of triangular subsets:

- Start with an equilateral triangle.
- Subdivide it into four smaller congruent equilateral triangles and remove the central triangle.
- Repeat step 2 with each of the remaining smaller triangles infinitely.

----
## Initial setup
We define the three edges of a triangle with equal sides, and store the coordinates in
the 2D array `points`.

We then plot this initial triangle, using the `fill` function from the `matplotlib` library.

```python
points = np.array([[-1,-0.5],[0,1],[1,-0.5]])
print(points)
print(points[:,0])
plt.fill(points[:,0],points[:,1],facecolor='blue', edgecolor='black')
```

----
## Create smaller triangles
We create the smaller triangles by halving the three sides, thus we get, in the first step,
four triangles, from which we remove the inner triangle.

We need two functions:
- one function `sirpinskiMidpoint` for halving the sides of each triangle
- one recursive function `sirpinskiTriangle`, drawing the outer three triangles for each triangle of each iteration

```python
def sierpinskiMidpoint(p1,p2):
    """
    Sierpinski midpoint
    calculate midpoint of a line (side of triangle)
    """
    xmid = (p1[0] + p2[0]) / 2
    ymid = (p1[1] + p2[1]) / 2
    return np.array([xmid,ymid])
```

```python
def sierpinskiTriangle(points,degree):
    """
    Sierpinski triangle 
    recursive function for plotting
    """
    #define color,map, blue for outer triangles, white for removed one
    colormap = ['blue']+['white' for i in range(degree)]
    # plot triangle
    ax.fill(points[:,0],points[:,1],facecolor=colormap[degree], edgecolor='black',linewidth=0)
    # recursively call sirpinski triangle generator
    if degree > 0:
        sierpinskiTriangle(np.vstack([points[0,:],
                   sierpinskiMidpoint(points[0,:], points[1,:]),
                   sierpinskiMidpoint(points[0,:], points[2,:])]),
                    degree-1)
        sierpinskiTriangle(np.vstack([points[1],
                    sierpinskiMidpoint(points[0], points[1]),
                    sierpinskiMidpoint(points[1], points[2])]),
                    degree-1)
        sierpinskiTriangle(np.vstack([points[2],
                    sierpinskiMidpoint(points[2], points[1]),
                    sierpinskiMidpoint(points[0], points[2])]),
                    degree-1)
    return
```

## Example for degree 1

```python
fig,ax=plt.subplots(figsize=(5,5)) 
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='both', which='both', length=0)
sierpinskiTriangle(points,1)
```

## Example for degree 2

```python
fig,ax=plt.subplots(figsize=(5,5)) 
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='both', which='both', length=0)
sierpinskiTriangle(points,2)
```

## Example for degree 10

```python
fig,ax=plt.subplots(figsize=(5,5))  
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='both', which='both', length=0)
sierpinskiTriangle(points,10)
```
