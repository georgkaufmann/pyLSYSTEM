```python
import numpy as np
import matplotlib.pyplot as plt
import json,csv
```

# Coastline data
Taken from [natural earth](https://www.naturalearthdata.com/downloads/50m-physical-vectors/)

```python
data = []
with open('data/GB.wkt', "r") as f:
    reader = csv.reader(f, delimiter="\t")
    for i, line in enumerate(reader):
        data.append(line[0])
        
data2 = ''.join(data)        
data3 = data2.split(',')

coord = np.empty((0,2))
print(coord.shape)
for i in range(len(data3)):
    lon = float(data3[i].split()[0])
    lat = float(data3[i].split()[1])
    #print(i,lon,lat)
    coord = np.append(coord,[[lon,lat]],axis=0)
    
print('min/max lon: ',coord[:,0].min(),coord[:,0].max())
print('min/max lat: ',coord[:,1].min(),coord[:,1].max())
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.set_aspect('equal','box')
ax.set_xlim([-7,2])
ax.set_ylim([50,59])
ax.fill(coord[:,0],coord[:,1],color=(0.5,0.5,0.5))
```
