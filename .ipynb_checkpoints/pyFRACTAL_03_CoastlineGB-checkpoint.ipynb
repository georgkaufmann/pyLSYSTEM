# Coastline data
We use a coastline data file from [natural earth](https://www.naturalearthdata.com/downloads/50m-physical-vectors/),
which provide the coastline as lontitude latitude data points, but in a weird format (wrt). 

We process this file as follows:
- Reading the file in with the `csv` library, we obtain a **single** line, stored as `data1`.
- We reformat this single line to a `str`, using the `join` command, and store the result in `data2`.
- We split this single string along the commata, using `split`, saving the results as list `data3`

```python
# read in coastline file
data1 = []
with open('data/GB.wkt', "r") as f:
    reader = csv.reader(f, delimiter="\t")
    for i, line in enumerate(reader):
        data1.append(line[0])
print('data1: ',type(data1),len(data1))

# get single lon/lat coordinates
data2 = ''.join(data1)
print('data2: ',type(data2),'\n',len(data2))

# split string to sub-strings
data3 = data2.split(',')
print('data3: ',type(data3),'\n',len(data3))
```

Next, we sort the coordinates into a 2D array `coordinates` ...

```python
# fill coordinate array
GBlonlat = np.empty((0,2))
print(coord.shape)
for i in range(len(data3)):
    lon = float(data3[i].split()[0])
    lat = float(data3[i].split()[1])
    #print(i,lon,lat)
    GBlonlat = np.append(GBlonlat,[[lon,lat]],axis=0)
print(GBlonlat.shape)
    
print('min/max lon: ',GBlonlat[:,0].min(),GBlonlat[:,0].max())
print('min/max lat: ',GBlonlat[:,1].min(),GBlonlat[:,1].max())
```

... and plot the coastline coordinates:

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.set_aspect('equal','box')
ax.set_xlim([-7,2])
ax.set_ylim([50,59])
ax.fill(GBlonlat[:,0],GBlonlat[:,1],color=(0.5,0.5,0.5))
```

## Convert to UTM
We convert the lon/lat data points to UTM coordinates, using the package [utm](https://pypi.org/project/utm) from pypi,
forcing zone 30.

We store the UTM coordinates in the 2D array `GButm`.

```python
import utm

GButm = np.zeros_like(GBlonlat)
print(GButm.shape)

for i in range(GBlonlat.shape[0]):
    GButm[i,0] = utm.from_latlon(GBlonlat[i,1],GBlonlat[i,0],force_zone_number=30)[0] / 1e3
    GButm[i,1] = utm.from_latlon(GBlonlat[i,1],GBlonlat[i,0],force_zone_number=30)[1] / 1e3


print(-GButm[:,0].min()+GButm[:,0].max())
print(-GButm[:,1].min()+GButm[:,1].max())
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
#ax.set_aspect('equal','box')
ax.set_xlim([300,900])
ax.set_ylim([5500,6600])
ax.fill(GButm[:,0],GButm[:,1],color=(0.5,0.5,0.5))
```

```python
# save UTM coordinates to file
f = open('data/GB.utm','w')
for i in range(GBlonlat.shape[0]):
    print("%10.2f %10.2f" % (GButm[i,0],GButm[i,1,]),file=f)
f.close()
```
